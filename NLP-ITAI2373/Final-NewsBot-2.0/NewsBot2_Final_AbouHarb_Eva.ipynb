```python
# 📊 Advanced Classification System

class AdvancedNewsClassifier:
    """Enhanced news classification with confidence scoring and explainability."""

    def __init__(self):
        self.vectorizer = TfidfVectorizer(stop_words='english', max_features=5000, ngram_range=(1, 2))
        self.model = LogisticRegression(max_iter=1000, class_weight='balanced')
        self.categories = []
        self.is_trained = False

    def train(self, X_train, y_train):
        """Train the classification model using article text and category labels."""
        self.categories = sorted(list(set(y_train)))
        X_vec = self.vectorizer.fit_transform(X_train)
        self.model.fit(X_vec, y_train)
        self.is_trained = True
        print(f'✅ Model trained on {len(self.categories)} categories: {self.categories}')

    def predict_with_confidence(self, article_text):
        """Predict article category with confidence scores and key phrase explanation."""
        if not self.is_trained:
            return {
                'primary_category': 'untrained',
                'confidence_score': 0.0,
                'all_scores': {},
                'explanation': 'Classifier must be trained before prediction.'
            }
        X_vec = self.vectorizer.transform([article_text])
        probabilities = self.model.predict_proba(X_vec)[0]
        prediction = self.model.classes_[int(np.argmax(probabilities))]
        category_scores = {cat: float(score) for cat, score in zip(self.model.classes_, probabilities)}
        return {
            'primary_category': prediction,
            'confidence_score': float(max(probabilities)),
            'all_scores': category_scores,
            'explanation': self.explain_prediction(article_text)
        }

    def explain_prediction(self, article_text):
        """Return top TF-IDF phrases that influenced the classification."""
        if not self.is_trained:
            return []
        feature_names = self.vectorizer.get_feature_names_out()
        X_vec = self.vectorizer.transform([article_text])
        nonzero_indices = X_vec.nonzero()[1]
        sorted_words = sorted(nonzero_indices, key=lambda i: X_vec[0, i], reverse=True)
        return [feature_names[i] for i in sorted_words[:5]]

# Demo training data keeps the notebook runnable even without an external dataset.
demo_articles = [
    'Apple released a new artificial intelligence chip for faster mobile computing.',
    'Microsoft announced cloud security tools for enterprise technology customers.',
    'The stock market rose after investors reacted to new inflation data.',
    'Banks reported stronger earnings while oil prices affected global markets.',
    'The local football team won the championship after a dramatic final match.',
    'The basketball coach praised players after the playoff victory.',
    'Lawmakers debated a new climate policy bill during the Senate session.',
    'The president announced international trade negotiations with European leaders.',
    'The film studio released a record-breaking movie with popular actors.',
    'A music festival attracted thousands of fans and entertainment reporters.'
]
demo_labels = [
    'technology', 'technology', 'business', 'business', 'sports', 'sports',
    'politics', 'politics', 'entertainment', 'entertainment'
]

newsbot.classifier = AdvancedNewsClassifier()
newsbot.classifier.train(demo_articles, demo_labels)

print('📊 Advanced classification system implemented!')
```

```python
# 🔍 Topic Modeling and Discovery

class TopicDiscoveryEngine:
    """Topic modeling engine for discovering themes and trends in article collections."""

    def __init__(self, n_topics=5):
        self.vectorizer = CountVectorizer(stop_words='english', max_features=2000)
        self.lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
        self.n_topics = n_topics
        self.is_fitted = False

    def fit_topics(self, documents):
        """Discover topics in a collection of documents."""
        if len(documents) < 2:
            raise ValueError('Topic modeling requires at least two documents.')
        X = self.vectorizer.fit_transform(documents)
        self.lda.fit(X)
        self.is_fitted = True
        print(f'✅ Discovered {self.n_topics} hidden topics in the dataset.')

    def get_article_topics(self, article_text):
        """Return the topic distribution for a single article."""
        if not self.is_fitted:
            return {'message': 'Topic model must be fitted first.'}
        X = self.vectorizer.transform([article_text])
        distribution = self.lda.transform(X)[0]
        dominant_topic = int(np.argmax(distribution))
        return {
            'dominant_topic': dominant_topic,
            'topic_distribution': [float(x) for x in distribution]
        }

    def extract_topics(self, documents=None, top_words=5):
        """Return readable topic labels using top words for each topic."""
        if documents is not None and not self.is_fitted:
            self.fit_topics(documents)
        if not self.is_fitted:
            return []
        feature_names = self.vectorizer.get_feature_names_out()
        topics = []
        for topic_idx, topic in enumerate(self.lda.components_):
            top_terms = [feature_names[i] for i in topic.argsort()[:-top_words - 1:-1]]
            topics.append({'topic_id': topic_idx, 'keywords': top_terms})
        return topics

    def track_topic_trends(self, articles_with_dates):
        """Summarize topic trends over time using article/date pairs or plain articles."""
        if not self.is_fitted:
            return {'message': 'Topic model must be fitted first.'}
        trend_counts = Counter()
        for item in articles_with_dates:
            text = item[0] if isinstance(item, tuple) else item
            topic = self.get_article_topics(text)['dominant_topic']
            trend_counts[topic] += 1
        return dict(trend_counts)

    def visualize_topics(self):
        """Return top topic words for simple visualization/reporting."""
        return self.extract_topics(top_words=6)

newsbot.topic_model = TopicDiscoveryEngine(n_topics=5)
newsbot.topic_model.fit_topics(demo_articles)

print('🔍 Topic discovery engine implemented!')
```

```python
# 🎭 Advanced Sentiment Analysis

class SentimentEvolutionTracker:
    """Sentiment analysis with polarity, subjectivity, VADER scores, and trend tracking."""

    def __init__(self):
        self.sia = SentimentIntensityAnalyzer()

    def analyze_sentiment(self, article_text):
        """Return overall sentiment, confidence, subjectivity, and emotion indicators."""
        if not article_text or not article_text.strip():
            return {'sentiment': 'neutral', 'confidence': 0.0, 'polarity': 0.0, 'subjectivity': 0.0, 'vader_scores': {}}
        blob = TextBlob(article_text)
        vader_scores = self.sia.polarity_scores(article_text)
        compound = vader_scores['compound']
        if compound >= 0.05:
            sentiment = 'positive'
        elif compound <= -0.05:
            sentiment = 'negative'
        else:
            sentiment = 'neutral'
        return {
            'sentiment': sentiment,
            'confidence': float(abs(compound)),
            'polarity': float(blob.sentiment.polarity),
            'subjectivity': float(blob.sentiment.subjectivity),
            'vader_scores': vader_scores,
            'key_phrases': self.extract_sentiment_phrases(article_text)
        }

    def extract_sentiment_phrases(self, article_text):
        """Identify sentences with the strongest sentiment signal."""
        sentences = sent_tokenize(article_text)
        scored = [(sent, abs(self.sia.polarity_scores(sent)['compound'])) for sent in sentences]
        scored.sort(key=lambda x: x[1], reverse=True)
        return [sent for sent, score in scored[:2] if score > 0]

    def track_sentiment_over_time(self, articles_with_dates):
        """Analyze average sentiment across dated or undated article collections."""
        rows = []
        for item in articles_with_dates:
            if isinstance(item, tuple):
                date, text = item
            else:
                date, text = 'unknown', item
            result = self.analyze_sentiment(text)
            rows.append({'date': date, 'sentiment': result['sentiment'], 'polarity': result['polarity'], 'compound': result['vader_scores'].get('compound', 0)})
        if not rows:
            return {'average_compound': 0, 'timeline': []}
        avg = sum(row['compound'] for row in rows) / len(rows)
        return {'average_compound': round(avg, 3), 'timeline': rows}

    def detect_sentiment_anomalies(self, sentiment_timeline):
        """Identify unusually strong positive or negative sentiment values."""
        values = []
        for item in sentiment_timeline:
            if isinstance(item, dict):
                values.append(item.get('compound', item.get('polarity', 0)))
            else:
                values.append(float(item))
        return [value for value in values if abs(value) >= 0.75]

newsbot.sentiment_tracker = SentimentEvolutionTracker()

print('🎭 Sentiment evolution tracker implemented!')
```

```python
# 🕸️ Entity Relationship Mapping

class EntityRelationshipMapper:
    """Named entity recognition and lightweight relationship mapping."""

    def __init__(self):
        self.nlp = nlp
        self.use_spacy = self.nlp is not None
        self.relationship_store = []

    def extract_entities(self, article_text):
        """Extract people, organizations, places, events, products, and dates."""
        entity_buckets = defaultdict(list)
        if self.use_spacy:
            doc = self.nlp(article_text)
            for ent in doc.ents:
                if ent.text not in entity_buckets[ent.label_]:
                    entity_buckets[ent.label_].append(ent.text)
        else:
            tokens = word_tokenize(article_text)
            tagged = nltk.pos_tag(tokens)
            for chunk in nltk.ne_chunk(tagged):
                if hasattr(chunk, 'label'):
                    entity = ' '.join(c[0] for c in chunk)
                    entity_buckets[chunk.label()].append(entity)
        return dict(entity_buckets)

    def extract_relationships(self, article_text):
        """Extract simple co-occurrence relationships within sentences."""
        relationships = []
        if self.use_spacy:
            doc = self.nlp(article_text)
            for sent in doc.sents:
                ents = [ent for ent in sent.ents if ent.label_ in ['PERSON', 'ORG', 'GPE', 'PRODUCT', 'EVENT']]
                for i in range(len(ents)):
                    for j in range(i + 1, len(ents)):
                        relation = {
                            'entity_1': ents[i].text,
                            'entity_1_type': ents[i].label_,
                            'entity_2': ents[j].text,
                            'entity_2_type': ents[j].label_,
                            'context': sent.text
                        }
                        relationships.append(relation)
                        self.relationship_store.append(relation)
        return relationships

    def build_knowledge_graph(self, articles):
        """Build a lightweight knowledge graph from extracted relationships."""
        graph = defaultdict(set)
        for article in articles:
            for rel in self.extract_relationships(article):
                graph[rel['entity_1']].add(rel['entity_2'])
                graph[rel['entity_2']].add(rel['entity_1'])
        return {node: sorted(list(neighbors)) for node, neighbors in graph.items()}

    def find_entity_connections(self, entity1, entity2):
        """Search stored relationships for a connection between two entities."""
        matches = []
        e1, e2 = entity1.lower(), entity2.lower()
        for rel in self.relationship_store:
            pair = {rel['entity_1'].lower(), rel['entity_2'].lower()}
            if e1 in pair and e2 in pair:
                matches.append(rel)
        return matches if matches else f'No direct stored connection found between {entity1} and {entity2}.'

newsbot.entity_mapper = EntityRelationshipMapper()

print('🕸️ Entity relationship mapper implemented!')
```

```python
# 📝 Intelligent Text Summarization

class IntelligentSummarizer:
    """Extractive text summarization with headline generation and quality checks."""

    def __init__(self):
        self.stop_words = set(stopwords.words('english'))

    def summarize_article(self, article_text, summary_type='balanced'):
        """Generate a readable summary using sentence scoring."""
        if not article_text or not article_text.strip():
            return ''
        sentence_count = {'brief': 2, 'balanced': 3, 'detailed': 5}.get(summary_type, 3)
        sentences = sent_tokenize(article_text)
        if len(sentences) <= sentence_count:
            return article_text
        words = [w.lower() for w in word_tokenize(article_text) if w.isalnum() and w.lower() not in self.stop_words]
        frequencies = Counter(words)
        sentence_scores = {}
        for sent in sentences:
            score = sum(frequencies.get(w.lower(), 0) for w in word_tokenize(sent) if w.isalnum())
            sentence_scores[sent] = score / max(len(word_tokenize(sent)), 1)
        top_sentences = heapq.nlargest(sentence_count, sentence_scores, key=sentence_scores.get)
        ordered = [sent for sent in sentences if sent in top_sentences]
        return ' '.join(ordered)

    def summarize_multiple_articles(self, articles, focus_topic=None):
        """Create a unified summary from multiple articles."""
        if focus_topic:
            articles = [a for a in articles if focus_topic.lower() in a.lower()]
        combined_text = ' '.join(articles)
        return self.summarize_article(combined_text, summary_type='detailed')

    def generate_headlines(self, article_text):
        """Generate simple headline candidates from important sentences."""
        summary = self.summarize_article(article_text, summary_type='brief')
        first_sentence = sent_tokenize(summary)[0] if sent_tokenize(summary) else article_text[:90]
        clean = first_sentence.strip().rstrip('.')
        return {
            'informative': clean[:100],
            'engaging': f'What to know: {clean[:85]}',
            'short': clean[:65]
        }

    def assess_summary_quality(self, original_text, summary):
        """Evaluate compression and coverage using simple transparent metrics."""
        original_words = max(len(original_text.split()), 1)
        summary_words = len(summary.split())
        overlap = len(set(original_text.lower().split()) & set(summary.lower().split()))
        return {
            'compression_ratio': round(summary_words / original_words, 3),
            'keyword_coverage': round(overlap / max(len(set(original_text.lower().split())), 1), 3),
            'summary_words': summary_words
        }

newsbot.summarizer = IntelligentSummarizer()

print('📝 Intelligent summarizer implemented!')
```

```python
# 🔍 Semantic Search and Similarity

class SemanticSearchEngine:
    """Semantic search using TF-IDF document vectors and cosine similarity."""

    def __init__(self):
        self.vectorizer = TfidfVectorizer(stop_words='english', max_features=5000, ngram_range=(1, 2))
        self.document_embeddings = None
        self.documents = []

    def encode_documents(self, documents):
        """Convert documents to vector embeddings."""
        self.documents = list(documents)
        self.document_embeddings = self.vectorizer.fit_transform(self.documents)
        return self.document_embeddings

    def find_similar_articles(self, query_article, top_k=5):
        """Find the most similar articles to a query article."""
        if self.document_embeddings is None:
            return []
        query_vec = self.vectorizer.transform([query_article])
        similarities = cosine_similarity(query_vec, self.document_embeddings).flatten()
        top_indices = similarities.argsort()[-top_k:][::-1]
        return [{'index': int(i), 'score': float(similarities[i]), 'article': self.documents[i]} for i in top_indices]

    def semantic_search(self, query_text, article_database=None, top_k=5):
        """Search articles using natural language meaning, not just exact keywords."""
        if self.document_embeddings is None:
            return []
        query_vec = self.vectorizer.transform([query_text])
        similarities = cosine_similarity(query_vec, self.document_embeddings).flatten()
        top_indices = similarities.argsort()[-top_k:][::-1]
        source = article_database if article_database is not None else self.documents
        results = []
        for i in top_indices:
            item = source.iloc[i].to_dict() if hasattr(source, 'iloc') else source[i]
            results.append({'index': int(i), 'score': float(similarities[i]), 'result': item})
        return results

    def cluster_similar_content(self, articles, n_clusters=3):
        """Group articles into clusters by semantic similarity."""
        if len(articles) < n_clusters:
            n_clusters = max(1, len(articles))
        X = self.vectorizer.fit_transform(articles)
        model = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
        labels = model.fit_predict(X)
        return [{'article': article, 'cluster': int(label)} for article, label in zip(articles, labels)]

newsbot.search_engine = SemanticSearchEngine()
newsbot.search_engine.encode_documents(demo_articles)

print('🔍 Semantic search engine implemented!')
```

```python
# 💡 Content Enhancement and Insights

class ContentEnhancer:
    """Content enhancement system for extracting context, facts, gaps, and insights."""

    def __init__(self):
        self.number_pattern = re.compile(r'\b\d+(?:\.\d+)?%?\b')

    def enhance_article(self, article_text):
        """Add context by extracting data points, entities, and readability indicators."""
        stats = self.number_pattern.findall(article_text)
        sentences = sent_tokenize(article_text) if article_text else []
        words = article_text.split()
        readability_note = 'short and easy to scan' if len(words) < 120 else 'detailed and information dense'
        return {
            'original_length_characters': len(article_text),
            'word_count': len(words),
            'sentence_count': len(sentences),
            'data_points': stats[:10],
            'readability_note': readability_note,
            'suggested_follow_up': 'Compare this article with related coverage for bias and missing context.'
        }

    def generate_insights(self, articles):
        """Generate high-level insights from an article collection."""
        combined = ' '.join(articles)
        common_terms = Counter(
            w.lower() for w in word_tokenize(combined)
            if w.isalpha() and len(w) > 4 and w.lower() not in stopwords.words('english')
        ).most_common(10)
        sentiment = newsbot.sentiment_tracker.track_sentiment_over_time(articles) if newsbot.sentiment_tracker else {}
        topics = newsbot.topic_model.extract_topics(articles) if newsbot.topic_model else []
        return {
            'article_count': len(articles),
            'top_terms': common_terms,
            'sentiment_trend': sentiment,
            'topics': topics[:5]
        }

    def detect_information_gaps(self, articles, topic):
        """Identify missing angles in coverage."""
        combined = ' '.join(articles).lower()
        expected_angles = ['history', 'impact', 'cost', 'people', 'policy', 'future']
        missing = [angle for angle in expected_angles if angle not in combined]
        return {
            'topic': topic,
            'missing_angles': missing,
            'recommendation': f'Add more background and stakeholder perspectives about {topic}.'
        }

    def cross_reference_facts(self, article_text):
        """Flag factual claims that should be verified with reliable external sources."""
        stats = self.number_pattern.findall(article_text)
        return {
            'claims_to_verify': stats,
            'verification_note': 'Numerical claims and named entities should be checked against trusted sources before publication.'
        }

newsbot.enhancer = ContentEnhancer()

print('💡 Content enhancer implemented!')
```

```python
# 🌐 Language Detection and Processing

class MultilingualProcessor:
    """Multilingual language detection, simple translation workflow, and cultural context checks."""

    def __init__(self):
        self.supported_languages = {'en': 'English', 'es': 'Spanish', 'fr': 'French', 'ar': 'Arabic', 'de': 'German'}

    def detect_language(self, text):
        """Detect language with confidence scoring."""
        try:
            results = detect_langs(text)
            return {
                'language': results[0].lang,
                'language_name': self.supported_languages.get(results[0].lang, 'Other/Unknown'),
                'confidence': float(results[0].prob),
                'all_detected': [str(r) for r in results]
            }
        except Exception:
            return {'language': 'unknown', 'language_name': 'Unknown', 'confidence': 0.0, 'all_detected': []}

    def translate_text(self, text, target_language='en'):
        """Translate or simulate translation while preserving transparency."""
        detected = self.detect_language(text)
        if detected['language'] == target_language:
            return {'translated_text': text, 'source_language': detected['language'], 'target_language': target_language, 'quality_note': 'No translation needed.'}
        # Offline-safe simulated translation. In production, connect this method to Google/Azure/DeepL API.
        return {
            'translated_text': f'[Translation placeholder from {detected["language"]} to {target_language}]: {text}',
            'source_language': detected['language'],
            'target_language': target_language,
            'quality_note': 'Offline-safe placeholder. Replace with a translation API for production deployment.'
        }

    def analyze_cross_lingual(self, articles_by_language):
        """Compare coverage volume and sentiment across languages."""
        report = {}
        for lang, articles in articles_by_language.items():
            sentiment = newsbot.sentiment_tracker.track_sentiment_over_time(articles) if newsbot.sentiment_tracker else {}
            report[lang] = {'article_count': len(articles), 'sentiment': sentiment}
        return report

    def extract_cultural_context(self, text, source_language):
        """Identify when cultural context review is especially important."""
        nuance_needed = {'zh', 'ar', 'ja', 'hi', 'ko'}
        return {
            'source_language': source_language,
            'cultural_context_level': 'high' if source_language in nuance_needed else 'standard',
            'note': 'Review idioms, regional politics, and local references when interpreting translated news.'
        }

newsbot.multilingual = MultilingualProcessor()

print('🌐 Multilingual processor implemented!')
```

```python
# 🎯 Intent Classification and Query Understanding

class ConversationalInterface:
    """Conversational AI layer for natural language interaction with NewsBot."""

    def __init__(self, newsbot_system):
        self.newsbot = newsbot_system
        self.context = {'last_intent': None, 'last_entities': None, 'history': []}

    def classify_intent(self, user_query):
        """Classify user intent from natural language query."""
        query = user_query.lower()
        if any(word in query for word in ['summarize', 'summary', 'brief']):
            return 'summarize'
        if any(word in query for word in ['sentiment', 'tone', 'feel', 'analyze']):
            return 'analyze'
        if any(word in query for word in ['compare', 'versus', 'vs']):
            return 'compare'
        if any(word in query for word in ['who is', 'relationship', 'connection', 'explain']):
            return 'explain'
        if any(word in query for word in ['translate', 'language', 'spanish', 'english']):
            return 'translate'
        if any(word in query for word in ['find', 'search', 'show me', 'articles about']):
            return 'search'
        return 'search'

    def extract_query_entities(self, user_query):
        """Extract categories, sentiment filters, languages, and comparison terms from a query."""
        query = user_query.lower()
        categories = ['technology', 'tech', 'business', 'sports', 'sport', 'entertainment', 'politics']
        sentiments = ['positive', 'negative', 'neutral']
        languages = {'spanish': 'es', 'english': 'en', 'french': 'fr', 'arabic': 'ar'}
        found_category = next((c for c in categories if c in query), None)
        if found_category == 'tech':
            found_category = 'technology'
        comparison_terms = re.split(r'\s+vs\s+|\s+versus\s+|\s+and\s+', user_query, flags=re.I)
        comparison_terms = [term.strip(' ?,.') for term in comparison_terms if len(term.strip()) > 2]
        return {
            'category': found_category,
            'sentiment': next((s for s in sentiments if s in query), None),
            'language': next((code for name, code in languages.items() if name in query), None),
            'comparison_terms': comparison_terms[:2] if len(comparison_terms) >= 2 else [],
            'raw_query': user_query
        }

    def process_query(self, user_query, conversation_context=None):
        """Process a natural language query and generate a response."""
        intent = self.classify_intent(user_query)
        entities = self.extract_query_entities(user_query)
        self.context['last_intent'] = intent
        self.context['last_entities'] = entities
        self.context['history'].append({'query': user_query, 'intent': intent, 'entities': entities})

        if intent == 'search' and self.newsbot.search_engine:
            query_results = self.newsbot.search_engine.semantic_search(user_query, top_k=3)
        elif intent == 'summarize' and self.newsbot.summarizer:
            query_results = self.newsbot.summarizer.summarize_multiple_articles(demo_articles, focus_topic=entities['category'])
        elif intent == 'analyze' and self.newsbot.sentiment_tracker:
            query_results = self.newsbot.sentiment_tracker.track_sentiment_over_time(demo_articles)
        elif intent == 'translate' and self.newsbot.multilingual:
            query_results = self.newsbot.multilingual.translate_text(user_query, target_language=entities['language'] or 'en')
        elif intent == 'explain' and self.newsbot.entity_mapper and len(entities['comparison_terms']) >= 2:
            query_results = self.newsbot.entity_mapper.find_entity_connections(entities['comparison_terms'][0], entities['comparison_terms'][1])
        else:
            query_results = 'Request understood. More article data would improve the answer.'
        return self.generate_response(query_results, intent, entities)

    def generate_response(self, query_results, intent, entities):
        """Generate helpful, natural language responses."""
        category_text = f' in {entities["category"]}' if entities.get('category') else ''
        if intent == 'search':
            return f'I found relevant news{category_text}. Top results: {query_results}'
        if intent == 'summarize':
            return f'Here is a concise summary{category_text}: {query_results}'
        if intent == 'analyze':
            return f'Sentiment analysis{category_text}: {query_results}'
        if intent == 'compare':
            return f'I can compare coverage using search, sentiment, and entity overlap. Extracted terms: {entities.get("comparison_terms")}'
        if intent == 'explain':
            return f'Entity relationship result: {query_results}'
        if intent == 'translate':
            return f'Translation workflow result: {query_results}'
        return 'I can search, summarize, analyze sentiment, translate, or explain entity relationships.'

    def handle_follow_up(self, follow_up_query, conversation_history=None):
        """Handle follow-up questions using stored context."""
        if self.context['last_entities'] and self.context['last_entities'].get('category'):
            follow_up_query += f' {self.context["last_entities"]["category"]}'
        return self.process_query(follow_up_query)

newsbot.conversation = ConversationalInterface(newsbot)

print('💬 Conversational interface implemented!')
```

```python
# 🔧 System Integration and Orchestration

class NewsBot2IntegratedSystem:
    """Complete NewsBot 2.0 system with all components integrated."""

    def __init__(self, config, base_system=None):
        self.config = config
        source = base_system if base_system is not None else newsbot
        self.classifier = source.classifier
        self.topic_engine = source.topic_model
        self.sentiment_tracker = source.sentiment_tracker
        self.entity_mapper = source.entity_mapper
        self.summarizer = source.summarizer
        self.search_engine = source.search_engine
        self.enhancer = source.enhancer
        self.multilingual = source.multilingual
        self.conversation = ConversationalInterface(self)
        self.processed_cache = {}

    def comprehensive_analysis(self, article_text):
        """Perform complete analysis of a single article."""
        if not article_text or not article_text.strip():
            return {'error': 'Article text is empty.'}
        cache_key = hash(article_text)
        if cache_key in self.processed_cache:
            return self.processed_cache[cache_key]
        lang_info = self.multilingual.detect_language(article_text)
        analysis_results = {
            'classification': self.classifier.predict_with_confidence(article_text),
            'sentiment': self.sentiment_tracker.analyze_sentiment(article_text),
            'entities': self.entity_mapper.extract_entities(article_text),
            'relationships': self.entity_mapper.extract_relationships(article_text),
            'topics': self.topic_engine.get_article_topics(article_text),
            'summary': self.summarizer.summarize_article(article_text),
            'headline_options': self.summarizer.generate_headlines(article_text),
            'enhancements': self.enhancer.enhance_article(article_text),
            'language': lang_info,
            'processed_at': datetime.now().isoformat()
        }
        self.processed_cache[cache_key] = analysis_results
        return analysis_results

    def batch_analysis(self, articles):
        """Analyze multiple articles with error handling."""
        results = []
        for idx, article in enumerate(articles):
            try:
                result = self.comprehensive_analysis(article)
                result['batch_index'] = idx
                results.append(result)
            except Exception as e:
                results.append({'batch_index': idx, 'error': str(e)})
        return results

    def query_interface(self, user_query):
        """Handle user queries through the conversational interface."""
        return self.conversation.process_query(user_query)

    def generate_insights_report(self, articles, report_type='comprehensive'):
        """Generate a report from multiple articles."""
        if report_type == 'summary':
            return self.summarizer.summarize_multiple_articles(articles)
        if report_type == 'trends':
            return {
                'sentiment_trends': self.sentiment_tracker.track_sentiment_over_time(articles),
                'topic_trends': self.topic_engine.track_topic_trends(articles)
            }
        return self.enhancer.generate_insights(articles)

newsbot2 = NewsBot2IntegratedSystem(config, newsbot)

print('🔧 Integrated system implemented!')
```

```python
# 🧪 Testing and Validation Framework

class NewsBot2TestSuite:
    """Comprehensive testing framework for NewsBot 2.0."""

    def __init__(self, newsbot_system):
        self.newsbot = newsbot_system

    def test_individual_components(self):
        """Test each component individually."""
        sample_text = 'Apple released a new iPhone in San Francisco today. Investors said the product launch was excellent.'
        tests = {}
        try:
            tests['classification'] = 'PASS' if self.newsbot.classifier.predict_with_confidence(sample_text)['primary_category'] else 'FAIL'
            tests['sentiment'] = 'PASS' if self.newsbot.sentiment_tracker.analyze_sentiment(sample_text)['sentiment'] else 'FAIL'
            tests['ner'] = 'PASS' if isinstance(self.newsbot.entity_mapper.extract_entities(sample_text), dict) else 'FAIL'
            tests['summarization'] = 'PASS' if self.newsbot.summarizer.summarize_article(sample_text) else 'FAIL'
            tests['translation'] = 'PASS' if self.newsbot.multilingual.detect_language(sample_text)['language'] else 'FAIL'
            tests['semantic_search'] = 'PASS' if self.newsbot.search_engine.semantic_search('technology products', top_k=1) else 'FAIL'
        except Exception as e:
            tests['error'] = str(e)
        return tests

    def test_integration(self):
        """Test integrated system functionality."""
        sample_text = 'Microsoft announced a new AI product for cloud customers. Analysts reacted positively to the technology update.'
        try:
            analysis = self.newsbot.comprehensive_analysis(sample_text)
            response = self.newsbot.query_interface('summarize technology news')
            return {'status': 'SUCCESS', 'analysis_keys': list(analysis.keys()), 'sample_response': response[:200]}
        except Exception as e:
            return {'status': 'FAILED', 'error': str(e)}

    def test_performance(self):
        """Test system speed on a sample article."""
        start = time.time()
        self.newsbot.comprehensive_analysis('Test article speed check about business and technology markets.')
        end = time.time()
        return {'processing_seconds': round(end - start, 4)}

    def test_edge_cases(self):
        """Test robustness with empty, short, and non-English text."""
        cases = {
            'empty': self.newsbot.comprehensive_analysis(''),
            'short': self.newsbot.comprehensive_analysis('Breaking news.'),
            'spanish': self.newsbot.multilingual.detect_language('El mercado tecnológico está creciendo rápidamente.')
        }
        return cases

    def run_all_tests(self):
        """Run every test and return one report."""
        return {
            'component_tests': self.test_individual_components(),
            'integration_test': self.test_integration(),
            'performance_test': self.test_performance(),
            'edge_case_tests': self.test_edge_cases()
        }

test_suite = NewsBot2TestSuite(newsbot2)
print('🧪 Testing framework implemented!')
print(test_suite.run_all_tests())
```

```python
# 📊 System Evaluation and Metrics

class NewsBot2Evaluator:
    """Comprehensive evaluation framework for NewsBot 2.0."""

    def __init__(self, newsbot_system):
        self.newsbot = newsbot_system

    def evaluate_classification_performance(self, test_data):
        """Evaluate classification accuracy, precision, recall, and F1-score."""
        if isinstance(test_data, pd.DataFrame):
            text_col = 'text' if 'text' in test_data.columns else test_data.columns[0]
            label_col = 'label' if 'label' in test_data.columns else test_data.columns[1]
            X_test = test_data[text_col].tolist()
            y_true = test_data[label_col].tolist()
        else:
            X_test, y_true = zip(*test_data)
        y_pred = [self.newsbot.classifier.predict_with_confidence(text)['primary_category'] for text in X_test]
        report = classification_report(y_true, y_pred, zero_division=0)
        matrix = confusion_matrix(y_true, y_pred)
        accuracy = accuracy_score(y_true, y_pred)
        print('--- Classification Performance ---')
        print(report)
        return {'accuracy': accuracy, 'report': report, 'confusion_matrix': matrix}

    def evaluate_topic_modeling_quality(self, documents):
        """Evaluate topic diversity and interpretability."""
        topics = self.newsbot.topic_engine.extract_topics(documents)
        all_keywords = [word for topic in topics for word in topic['keywords']]
        diversity_score = len(set(all_keywords)) / max(len(all_keywords), 1)
        return {'topic_count': len(topics), 'topic_diversity_score': round(diversity_score, 3), 'topics': topics}

    def evaluate_summarization_quality(self, articles_and_summaries):
        """Evaluate summarization compression and keyword coverage."""
        return [self.newsbot.summarizer.assess_summary_quality(article, summary) for article, summary in articles_and_summaries]

    def evaluate_user_experience(self, user_interactions):
        """Evaluate query understanding coverage."""
        intents = [self.newsbot.conversation.classify_intent(q) for q in user_interactions]
        recognized = [intent for intent in intents if intent in ['search', 'summarize', 'analyze', 'compare', 'explain', 'translate']]
        return {
            'queries_tested': len(user_interactions),
            'recognized_intents': intents,
            'task_completion_rate': round(len(recognized) / max(len(user_interactions), 1), 3)
        }

    def generate_evaluation_report(self):
        """Generate a comprehensive evaluation report."""
        sample_summary = self.newsbot.summarizer.summarize_article(demo_articles[0])
        summary_quality = self.newsbot.summarizer.assess_summary_quality(demo_articles[0], sample_summary)
        topic_quality = self.evaluate_topic_modeling_quality(demo_articles)
        ux_quality = self.evaluate_user_experience([
            'search technology news',
            'summarize business articles',
            'analyze sentiment in politics',
            'translate this to Spanish'
        ])
        return {
            'system_name': 'NewsBot 2.0',
            'evaluation_date': datetime.now().isoformat(),
            'summary_quality': summary_quality,
            'topic_quality': topic_quality,
            'user_experience': ux_quality,
            'strengths': [
                'Integrated classification, sentiment, topic modeling, NER, summarization, search, and multilingual detection.',
                'Offline-safe demo dataset makes the notebook runnable without external APIs.',
                'Modular design supports production expansion.'
            ],
            'limitations': [
                'Translation uses an offline-safe placeholder unless connected to a production API.',
                'Classifier accuracy should be re-evaluated using a larger real news dataset.'
            ],
            'recommendations': [
                'Connect to a real news API.',
                'Add a Streamlit frontend for bonus points.',
                'Fine-tune transformer models for higher-quality summarization and classification.'
            ]
        }

evaluator = NewsBot2Evaluator(newsbot2)
print('📊 Evaluation framework implemented!')
evaluator.generate_evaluation_report()
```
